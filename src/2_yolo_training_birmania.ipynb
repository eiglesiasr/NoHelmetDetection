```python
from ultralytics import YOLO
import os
from IPython.display import Image, display

dataset_path = '../dataset'
data_yaml = os.path.join(dataset_path, 'databirmania.yaml')
```

```python
# Carga modelo base (puedes usar yolov8n.pt, yolov8s.pt, etc.)
model = YOLO('yolov8n.pt')  # o yolov8s.pt según la capacidad

# Entrena (fine-tune)
model.train(
    data=data_yaml,
    epochs=50,
    imgsz=640,
    batch=16,
    project='noHelmetDetection',
    name='exp3',
    exist_ok=True
)
```

```python
# cargar modelo
model = YOLO('noHelmetDetection/exp3/weights/best.pt')
```

```python
metrics_test = model.val(data=data_yaml, split="val", imgsz=640)
```

```python
# Predice sobre una imagen de prueba
results = model.predict(source=f"{dataset_path}/birmania/images/val/bago_highway_1_001.jpg", save=True)
```
